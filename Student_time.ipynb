```python
import matplotlib.pyplot as plt
import pandas as pd

from google.colab import files
uploaded = files.upload()

import io
data =  pd.read_csv(io.BytesIO(uploaded['Student_time_analyze.csv']))

plt.plot(data.student_ID,data.start_TIME,color='b')
plt.xlabel('StudentID')
plt.ylabel('Time')
plt.xticks(data.student_ID,rotation='vertical')
plt.title('first anser time')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

#from google.colab import files
#uploaded = files.upload()

import io
data =  pd.read_csv(io.BytesIO(uploaded['Student_time_analyze.csv']))

plt.plot(data.student_ID,data.start_TIME,color='b')
plt.plot(data.student_ID,data.finish_TIME,color='r')
plt.xlabel('StudentID')
plt.ylabel('Time')
plt.xticks(data.student_ID,rotation='vertical')
plt.title('Total(not sort)')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

#from google.colab import files
#uploaded = files.upload()

import io
data = pd.read_csv(io.BytesIO(uploaded['Student_time_analyze.csv']))

takeTime = data.finish_TIME - data.start_TIME
plt.plot(data.student_ID,takeTime,color='r')
plt.xlabel('StudentID')
plt.ylabel('Time')
plt.xticks(data.student_ID,rotation='vertical')
plt.title('Total(not sort)')
plt.show()
```
